```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.lines import Line2D


# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

#Reading csvs
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combining DataFrames 
Merged_df = pd.merge(city_data, ride_data, on="city")
Merged_df.head()


```

```python
Merged_df.columns
```

```python

Merged_df['type'] = Merged_df['type'].replace({'Urban': 1, 'Rural': 2, 'Suburban': 3})
```

```python
# Average Fare per City. Y-axis 
avg_fare = Merged_df.groupby(['city']).mean()['fare']


# Rides Per city. X-axis  
ride_count = Merged_df.groupby(['city']).count()['date']


# Radius - Driver Count per City
driver_count = Merged_df.groupby(['city']).mean()['driver_count']

# Defining Colors
colors = Merged_df.groupby('city').mean()['type']
new_colors = colors.replace({1: "Coral", 2: "gold", 3: "SkyBlue"})




```

```python
plt.scatter(ride_count, avg_fare, s=(driver_count*10) , marker="o", facecolors=new_colors,
             edgecolors="black", alpha=0.75)

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid(b=None, which='major', axis='both',)

legend_elements = [Line2D([0], [0], marker='o', color='w', label='Urban', markerfacecolor='Coral', markersize=10),
                  Line2D([0], [0], marker='o', color='w', label='Suburban', markerfacecolor='Skyblue', markersize=10),
                  Line2D([0], [0], marker='o', color='w', label='Rural', markerfacecolor='gold', markersize=10)]

plt.legend(handles=legend_elements,)                   
plt.show
```

```python
# Pie Chart (Total Fares by City Type)
Merged_pie_df = pd.merge(city_data, ride_data, on="city")
#Merged_pie_df.head(2)
```

```python
# Revanue per City Type
fare_by_type = Merged_pie_df.groupby(['type']).sum()['fare']

# Total Fare Revanue
sum_fare = Merged_pie_df.sum()['fare']

#Determining Slices of the pie 
slices = round((fare_by_type/sum_fare)*100,2)

# Pie Elements
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "SkyBlue", "Coral"]
explode = (0, 0, 0.15)
# Making Pie 
plt.pie(slices, labels=labels, explode=explode, startangle=140, colors=colors, autopct="%1.1f%%")
plt.title("Revanue Per City Type")
plt.axis("equal")
plt.show
```

```python
# Total Rides by City Type
Rides_Per_City = Merged_pie_df.groupby(['type']).count()['ride_id']
Rides_Per_City

plt.pie(Rides_Per_City, labels=labels, explode=explode, startangle=140, colors=colors, autopct="%1.1f%%")
plt.title("Total Rides by City Type")
plt.axis("equal")
plt.show
```

```python
# Total Drivers by City Type
Unique_city_Merged = Merged_pie_df.drop_duplicates('city')
Unique_city_Merged

drivers_by_type = Unique_city_Merged.groupby(['type']).sum()['driver_count']
drivers_by_type

plt.pie(drivers_by_type, labels=labels, explode=explode, startangle=140, colors=colors, autopct="%1.1f%%")
plt.title("Total Drivers by City Type")
plt.axis("equal")
plt.show
```

```python
# Three Observations about the data

# 1) There is an inverse relationship for the price of a ride and the total number of rides per city. 
# 2) Urban cities have more dirvers, rides, and Revanue than Suburban & Rural cities combined. 
# 3) Rural cities rides seem to have the most expensive fares, where Urban city rides have to least expensive rides. 
```
